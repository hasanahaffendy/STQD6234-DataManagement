<h1><strong>Customer Shopping Trends </h1></strong>

<h1><strong>Introduction </h1></strong>


The Customer Shopping Preferences Dataset provides businesses with valuable insights into consumer behavior and purchasing patterns. In today's competitive market, understanding customer preferences and trends is critical for customizing products, marketing strategies, and the overall customer experience. Businesses can use this dataset to gain insights into a variety of customer attributes, such as age, gender, purchase history, preferred payment methods, and more. These insights help businesses make more informed decisions, optimize product offerings, and ultimately increase customer satisfaction. With 3900 records, this dataset establishes a solid foundation for data-driven decision-making and customer-centric strategies.



<h1><strong>Key Features: </h1></strong>

1. <strong>Customer ID</strong>: Unique identifier for each customer
2. <strong>Age</strong>: Age of the customer
3. <strong>Gender</strong>: Gender of the customer (Male/Female)
4. <strong>Item Purchased</strong>: The item purchased by the customer
5. <strong>Category</strong>: Category of the item purchased
6. <strong>Purchase Amount (USD)</strong>: The amount of the purchase in USD
7. <strong>Location</strong>: Location where the purchase was made
8. <strong>Size</strong>: Size of the purchased item
9. <strong>Color</strong>: Color of the purchased item
10. <strong>Season</strong>: Season during which the purchase was made
11. <strong>Review Rating</strong>: Rating given by the customer for the purchased item
12. <strong>Subscription Status</strong>: Indicates if the customer has a subscription (Yes/No)
13. <strong>Shipping Type</strong>: Type of shipping chosen by the customer
14. <strong>Discount Applied</strong>: Indicates if a discount was applied to the purchase (Yes/No)
15. <strong>Promo Code Used</strong>: Indicates if a promo code was used for the purchase (Yes/No)
16. <strong>Previous Purchases</strong>: The total count of transactions concluded by the customer at the store, excluding the ongoing transaction
17. <strong>Payment Method</strong>: Customer's most preferred payment method
18. <strong>Frequency of Purchases</strong>: Frequency at which the customer makes purchases (e.g., Weekly, Fortnightly, Monthly)



Link: https://www.kaggle.com/datasets/iamsouravbanerjee/customer-shopping-trends-dataset/data


<h1><strong>Data Cleaning </h1></strong>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Upload a single file
df = pd.read_csv('/content/drive/MyDrive/DataScience/shopping_trends_updated.csv')
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values:")
print(missing_values)
```

```python
# Check for duplicates
duplicates = df.duplicated()
print("Duplicate rows:")
print(duplicates)
```

```python
# Count the number of duplicate rows
print("Number of duplicate rows:", duplicates.sum())
```

```python
# Check for outliers in age below 18 and above 70
outliers_below_18 = df[df['Age'] < 18]
outliers_above_70 = df[df['Age'] > 70]

print("Outliers below 18:")
print(outliers_below_18)

print("\nOutliers above 70:")
print(outliers_above_70)
```

```python
# Check for review ratings outside the expected range
outliers_below_2_5 = df[df['Review Rating'] < 2.5]
outliers_above_5_0 = df[df['Review Rating'] > 5.0]

print("Outliers below 2.5:")
print(outliers_below_2_5)

print("\nOutliers above 5.0:")
print(outliers_above_5_0)
```

<h1><strong>Data Visualization </h1></strong>

```python
# Plot distribution of purchase amounts
plt.figure(figsize=(10, 6))
sns.histplot(df['Purchase Amount (USD)'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Purchase Amounts')
plt.xlabel('Purchase Amount (USD)')
plt.ylabel('Frequency')
plt.grid(True)

# Add statistical summary
plt.axvline(df['Purchase Amount (USD)'].mean(), color='red', linestyle='--', label='Mean')
plt.axvline(df['Purchase Amount (USD)'].median(), color='green', linestyle='--', label='Median')
plt.legend()

# Annotate mean and median values
plt.text(df['Purchase Amount (USD)'].mean() + 2, 200, f'Mean: ${df["Purchase Amount (USD)"].mean():.2f}', color='red')
plt.text(df['Purchase Amount (USD)'].median() + 2, 205, f'Median: ${df["Purchase Amount (USD)"].median():.2f}', color='green')

# Annotate the difference between mean and median
difference = df['Purchase Amount (USD)'].mean() - df['Purchase Amount (USD)'].median()
plt.text(df['Purchase Amount (USD)'].median() + 2, 210, f'Difference: ${difference:.2f}', color='blue')

plt.show()
```

Examining the histogram for the distribution of purchase amounts, we can see that the mean purchase amount is approximately 59.76 USD, while the median purchase amount is 60.00 USD. Interestingly, the mean and median values differ by only 0.24 USD, indicating a relatively symmetric distribution. This small difference indicates that the majority of purchase amounts are clustered around the median value, with a few extreme values pulling the mean slightly lower. As a result, the distribution of purchase amounts is nearly symmetrical, with the median serving as a reliable measure of central tendency in this dataset.


```python
# Plot relationship between age and purchase amount
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Purchase Amount (USD)', data=df, color='coral', alpha=0.6)
plt.title('Age vs. Purchase Amount')
plt.xlabel('Age')
plt.ylabel('Purchase Amount (USD)')
plt.grid(True)

# Add a trend line
sns.regplot(x='Age', y='Purchase Amount (USD)', data=df, scatter=False, color='blue')

plt.show()
```

The scatter plot depicting the relationship between age and purchase amount shows a clear trend as age increases. A trend line, which is a straight line spanning from age 18 to 70, depicts the overall pattern of purchase amounts across age groups. Interestingly, despite variations in individual purchasing behavior, the trend line has remained relatively stable, hovering around $60 for most age groups. This suggests that spending patterns remain consistent across a wide age range. The trend line's straight trajectory emphasizes the absence of significant age-related fluctuations in purchase amounts, highlighting the dataset's stable consumer spending trend.


```python
# Plot count of purchases by gender
plt.figure(figsize=(8, 5))
ax = sns.countplot(x='Gender', data=df, palette='pastel', hue='Gender', legend=False)
plt.title('Count of Purchases by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(axis='y')

# Add percentage labels
total_count = len(df)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total_count)
    x = p.get_x() + p.get_width() / 2 - 0.1
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom', fontsize=10)

plt.show()
```

The number of purchases by gender provides intriguing insights into consumer behavior. The graph shows a notable difference in purchasing frequency between men and women. With over 2500 purchases attributed to males, they account for roughly 68% of total purchases. Females, on the other hand, make between 1000 and 1300 purchases, accounting for roughly 32% of total purchases. This significant difference suggests that males are more likely to engage in purchasing activities than females. Understanding gender differences in purchasing behavior is critical for businesses to tailor their marketing strategies and product offerings to effectively target and cater to different consumer segments.


```python
# Plot distribution of review ratings
plt.figure(figsize=(8, 5))
sns.histplot(df['Review Rating'], bins=5, discrete=True, color='green')
plt.title('Distribution of Review Ratings')
plt.xlabel('Review Rating')
plt.ylabel('Frequency')
plt.grid(True)

# Add vertical lines for mean and median review ratings
mean_rating = df['Review Rating'].mean()
median_rating = df['Review Rating'].median()

plt.axvline(mean_rating, color='red', linestyle='--', label=f'Mean: {mean_rating:.2f}')
plt.axvline(median_rating, color='blue', linestyle='--', label=f'Median: {median_rating:.2f}')
plt.legend()

plt.show()
```

The histogram of the distribution of review ratings shows that ratings ranging from 3.0 to 5.0 have the highest frequency, indicating that the majority of reviews fall into this range. The mean review rating is calculated to be 3.75, and the median review rating is 3.7. This suggests that the majority of reviews are positive, with ratings clustering near the top of the scale. Understanding the central tendency of review ratings is crucial for evaluating overall customer satisfaction and identifying areas for improvement.

```python
# Plot boxplot of purchase amount by gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Purchase Amount (USD)', data=df, palette='Set2')
plt.title('Purchase Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Purchase Amount (USD)')
plt.grid(True)

# Add mean markers
mean_purchase_male = df[df['Gender'] == 'Male']['Purchase Amount (USD)'].mean()
mean_purchase_female = df[df['Gender'] == 'Female']['Purchase Amount (USD)'].mean()

plt.scatter(x=0, y=mean_purchase_male, color='red', marker='o', s=100, label=f'Mean Male: ${mean_purchase_male:.2f}')
plt.scatter(x=1, y=mean_purchase_female, color='blue', marker='o', s=100, label=f'Mean Female: ${mean_purchase_female:.2f}')
plt.legend()

plt.show()
```

The box plot depicting purchase amount by gender shows that the mean purchase amount for males is approximately $59.94 USD, while females spend around $60.25 USD. This slight difference implies that, on average, females spend slightly more than males. Understanding the gender differences in purchasing behavior can help businesses tailor their marketing strategies and product offerings accordingly.



```python
# Define the order of purchase frequencies
purchase_order = ['Fortnightly', 'Weekly', 'Bi-Weekly', 'Monthly', 'Every 3 Months', 'Quarterly', 'Annually']

# Plot bar chart of purchase frequency by subscription status
plt.figure(figsize=(10, 6))
sns.countplot(x='Frequency of Purchases', hue='Subscription Status', data=df, palette='muted', order=purchase_order)
plt.title('Purchase Frequency by Subscription Status')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Count')
plt.legend(title='Subscription Status')
plt.grid(axis='y')

plt.show()
```

The bar chart shows that customers without subscriptions make more purchases than those who have subscriptions. Customers without subscriptions, in particular, have consistently higher purchase counts across a range of purchase frequencies.

Customers with subscriptions have lower purchase counts, with less than 200 counts recorded for fortnightly, weekly, bi-weekly, monthly, every three months, quarterly, and  annually. Customers without subscriptions, on the other hand, have purchase counts that exceed 400 for the majority of purchase frequencies, with the exception of fortnightly and weekly, which is slightly less than 400.

This disparity in purchase behavior emphasizes the potential impact of subscription status on customer purchasing habits. Businesses can use this information to tailor their marketing strategies and subscription offerings, increasing customer engagement and encouraging repeat purchases.


```python
# Plot pie chart of payment method distribution
payment_method_counts = df['Payment Method'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(payment_method_counts, labels=payment_method_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Payment Method Distribution')
plt.axis('equal')
plt.show()
```

According to the pie chart, the majority of customers use PayPal, which accounts for approximately 17.4% of all payment methods. Following PayPal, credit cards and cash are tied for second place, accounting for 17.2% of total transactions. Debit cards and Venmo are tied for third place, each accounting for 16.3% of the total. Finally, bank transfers are the least popular payment method, accounting for 15.7% of the total.

This distribution provides insights into customers' preferred payment methods, which can help businesses optimize their payment processing systems and provide convenient payment options to customers.


```python
# Plot count of purchases by category
plt.figure(figsize=(12, 6))
sns.countplot(x='Category', data=df, palette='Set3', order=df['Category'].value_counts().index)
plt.title('Count of Purchases by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.grid(axis='y')

plt.show()
```

The bar chart shows that customers buy clothing the most, with nearly 1750 counts. Following clothing, accessories are the next most popular category, with approximately 1250 purchases. Outwear is the least purchased category, with counts ranging from 250 to 300, while footwear falls between 500 and 550.

This breakdown of purchases by category reveals important information about customer preferences, which can help businesses tailor their inventory and marketing strategies to better meet customer demand and maximize profits.


```python
# Plot distribution of purchase amounts by category
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Purchase Amount (USD)', data=df, palette='viridis', order=df['Category'].value_counts().index)
plt.title('Distribution of Purchase Amounts by Category')
plt.xlabel('Category')
plt.ylabel('Purchase Amount (USD)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)

# Add mean and median annotations for each category
grouped = df.groupby('Category')['Purchase Amount (USD)']
for i, (cat, grp) in enumerate(grouped):
    plt.text(i, grp.median(), f'Median: ${grp.median():.2f}\nMean: ${grp.mean():.2f}', ha='center', va='center', bbox=dict(facecolor='white', alpha=0.5))

plt.show()
```

In the case of clothing, accessories, and footwear, although they have varying purchase frequencies, their mean and median purchase amounts are quite similar. This suggests that while these categories may have different purchase frequencies, the average amount spent on purchases within each category is relatively consistent.

On the other hand, outwear stands out with a significantly lower mean purchase amount compared to the other categories. This indicates that while outwear may not be as frequently purchased as clothing or accessories, customers tend to spend less on individual outwear purchases.

These insights can help businesses tailor their marketing strategies and inventory management approaches to maximize sales and better meet customer preferences within each category.

```python
# Plot correlation heatmap of numerical features
numerical_features = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_features.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

It is interesting to note that the correlation heatmap shows relatively low correlations between most variables, with the highest correlation coefficient of 0.04 between age and previous purchases. This indicates that there is no strong linear relationship between these variables.

Furthermore, the correlation coefficient of 0.03 between review rating and purchase amount (USD) suggests a weak positive relationship. This implies that there may be a slight correlation between higher review ratings and slightly higher purchase amounts, but the relationship is not statistically significant.

Furthermore, 0.01 correlations between purchase amount (USD) and customer ID, as well as between purchase amount (USD) and previous purchases, indicate that these variables have very weak or negligible linear relationships with one another.
