<h1><strong>Customer Shopping Trends </h1></strong>

<h1><strong>Introduction </h1></strong>


The Customer Shopping Preferences Dataset provides businesses with valuable insights into consumer behavior and purchasing patterns. In today's competitive market, understanding customer preferences and trends is critical for customizing products, marketing strategies, and the overall customer experience. Businesses can use this dataset to gain insights into a variety of customer attributes, such as age, gender, purchase history, preferred payment methods, and more. These insights help businesses make more informed decisions, optimize product offerings, and ultimately increase customer satisfaction. With 3900 records, this dataset establishes a solid foundation for data-driven decision-making and customer-centric strategies.



<h1><strong>Key Features: </h1></strong>

1. <strong>Customer ID</strong>: Unique identifier for each customer
2. <strong>Age</strong>: Age of the customer
3. <strong>Gender</strong>: Gender of the customer (Male/Female)
4. <strong>Item Purchased</strong>: The item purchased by the customer
5. <strong>Category</strong>: Category of the item purchased
6. <strong>Purchase Amount (USD)</strong>: The amount of the purchase in USD
7. <strong>Location</strong>: Location where the purchase was made
8. <strong>Size</strong>: Size of the purchased item
9. <strong>Color</strong>: Color of the purchased item
10. <strong>Season</strong>: Season during which the purchase was made
11. <strong>Review Rating</strong>: Rating given by the customer for the purchased item
12. <strong>Subscription Status</strong>: Indicates if the customer has a subscription (Yes/No)
13. <strong>Shipping Type</strong>: Type of shipping chosen by the customer
14. <strong>Discount Applied</strong>: Indicates if a discount was applied to the purchase (Yes/No)
15. <strong>Promo Code Used</strong>: Indicates if a promo code was used for the purchase (Yes/No)
16. <strong>Previous Purchases</strong>: The total count of transactions concluded by the customer at the store, excluding the ongoing transaction
17. <strong>Payment Method</strong>: Customer's most preferred payment method
18. <strong>Frequency of Purchases</strong>: Frequency at which the customer makes purchases (e.g., Weekly, Fortnightly, Monthly)

<h1><strong>Data Cleaning </h1></strong>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Upload a single file
df = pd.read_csv('/content/drive/MyDrive/DataScience/shopping_trends_updated.csv')
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values:")
print(missing_values)
```

```python
# Check for duplicates
duplicates = df.duplicated()
print("Duplicate rows:")
print(duplicates)
```

```python
# Plot distribution of purchase amounts
plt.figure(figsize=(10, 6))
sns.histplot(df['Purchase Amount (USD)'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Purchase Amounts')
plt.xlabel('Purchase Amount (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Plot relationship between age and purchase amount
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Purchase Amount (USD)', data=df, color='coral', alpha=0.6)
plt.title('Age vs. Purchase Amount')
plt.xlabel('Age')
plt.ylabel('Purchase Amount (USD)')
plt.grid(True)
plt.show()
```

```python
# Plot count of purchases by gender
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Count of Purchases by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()
```

```python
# Plot distribution of review ratings
plt.figure(figsize=(8, 5))
sns.histplot(df['Review Rating'], bins=5, discrete=True, color='green')
plt.title('Distribution of Review Ratings')
plt.xlabel('Review Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Plot boxplot of purchase amount by gender
plt.figure(figsize=(8, 5))
sns.boxplot(x='Gender', y='Purchase Amount (USD)', data=df, palette='Set2')
plt.title('Purchase Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Purchase Amount (USD)')
plt.grid(True)
plt.show()
```

```python
# Plot bar chart of purchase frequency by subscription status
plt.figure(figsize=(8, 5))
sns.countplot(x='Frequency of Purchases', hue='Subscription Status', data=df, palette='muted')
plt.title('Purchase Frequency by Subscription Status')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Count')
plt.legend(title='Subscription Status')
plt.grid(axis='y')
plt.show()
```

```python
# Plot pie chart of payment method distribution
payment_method_counts = df['Payment Method'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(payment_method_counts, labels=payment_method_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Payment Method Distribution')
plt.axis('equal')
plt.show()
```

```python
# Plot count of purchases by category
plt.figure(figsize=(12, 6))
sns.countplot(x='Category', data=df, palette='Set3')
plt.title('Count of Purchases by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Plot distribution of purchase amounts by category
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Purchase Amount (USD)', data=df, palette='viridis')
plt.title('Distribution of Purchase Amounts by Category')
plt.xlabel('Category')
plt.ylabel('Purchase Amount (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Plot correlation heatmap of numerical features
numerical_features = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_features.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
